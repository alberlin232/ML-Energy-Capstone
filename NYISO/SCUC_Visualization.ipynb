```python
width = .35

fig, ax = plt.subplots()

ax.bar(df['Row'], df['Rnw Gen(MW)'], width, label='Rnw Gen(MW)')
ax.bar(df['Row'], df['Tot Demand(MW)']-df['Rnw Gen(MW)'], width, bottom=df['Rnw Gen(MW)'], label='Tot Demand(MW)')

ax.set_ylabel('MW')
ax.set_title('Total Demand over Renewable Generation Weekly')
plt.ylim(0,df['Tot Demand(MW)'].max()+150000)
ax.legend()

plt.show()
```

```python
color_map1 = ["#CC0000", "#0000CC"]
color_map2 = ["#FFC026", "#CCCCCC"]
color_map22 = ["#CCCCCC", "#FFC026"]
color_map3 = ["#34495E", "#9b59b6"]
color_map4 = ["#008000", "#f9d71c"]
fig, ax = plt.subplots()
plt.stackplot(df['Row'],df['Rnw Gen(MW)'], df['Tot Demand(MW)']-df['Rnw Gen(MW)'], labels=['Rnw Gen(MW)','Tot Demand(MW)'], colors=color_map2)
ax.set_ylabel('MW')
ax.set_title('Total Demand over Renewable Generation Weekly')
plt.ylim(0,df['Tot Demand(MW)'].max()+150000)
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
plt.stackplot(df['Row'],df['Disp Cost($)'], df['Res Cost($)'], labels=['Disp Cost($)','Res Cost($)'], colors=color_map22)
ax.set_ylabel('Dollars')
ax.set_title('Total Cost Weekly')
plt.ylim(0,df['Tot Cost($)'].max()+2000000)
plt.legend()
plt.show()
```
